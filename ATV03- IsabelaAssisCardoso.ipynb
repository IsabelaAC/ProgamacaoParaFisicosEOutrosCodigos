```python
# Atividade 1 - Ajuste a interpolação  de uma curva do tipo : y = a * sin(b*x + c) usando 
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline";
# Faça a discussão dos resultados
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(-2*np.pi, 2*np.pi, 50)

def f(x, a, b, c):
    return a*np.sin(b*x + c)

y = f(x, 1, 1, 0)

ruido = 0.3*np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.85)

xn = np.linspace(-2*np.pi, 2*np.pi, 100 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()



#Discussão: Foi plotado os pontos da função f(x)= sen(x) com ruído criado artificialmente. A linha alaranjada
#representa a interpolação linear para estes pontos, a linha verde tracejada a interpolação cúbica e a linha
#vermelha tracejada a interpolação do tipo "Cubic spline". Pode-se ver que, no intervalo apresentado, a interpola-
#ção que mais se aproxima da função sen(x) é a linha vermelha, não tendo tantas oscilações como as outras duas e 
#se aproximando mais da forma de onda, logo, sendo a interpolação mais adequada para esta função.
```

```python
## Atividade 2 - Ajuste a interpolação  de uma curva do tipo :  usando y = a/( (x - b)**2 + c )
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline"; 
# Faça a discussão dos resultados

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(0.1, 5, 50)

def f(x, a, b, c):
    return a/( (x - b)**2 + c )

y = f(x, 1, 0, 0)

ruido = np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.75)

xn = np.linspace(0.1, 5, 100 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()

#Discussão: Neste caso, a função plotada foi f(x) = 1/x^2, e, como pode-se ver, todos os 3 métodos se mostraram efi-
#cazes para interpolar a função, se aproximando do comportamento desejado, com pequenas ondulações, não havendo gran-
#des diferenças perceptíveis para este caso.
```

```python
## Atividade 3 - Ajuste a interpolação  de uma curva do tipo :  usando y = a* sin(b*x)*sqrt(x+c)
# (a) interpolação polinomial linear e cúbica;
# (b) interpolação tipo "cubic spline"; 
# Faça a discussão dos resultados

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.interpolate import UnivariateSpline
from scipy import interpolate


x = np.linspace(0.1, 20, 100)

def f(x, a, b, c):
    return a*np.sin(b*x)*np.sqrt(x+c)

y = f(x, 1, 1, 0)

ruido = 0.3*np.random.normal(size = y.size)

y += ruido

inter1 = interpolate.interp1d(x, y, kind = 'linear')

inter3 = interpolate.interp1d(x, y, kind = 'cubic')

spl = UnivariateSpline(x, y)

spl.set_smoothing_factor(1.75)

xn = np.linspace(0.1, 20, 200 )

yn1 = inter1(xn)
yn2 = inter3(xn)
yn3 = spl(xn)

plt.plot(x, y, 'o', xn, yn1, '-', xn, yn2, '--', xn, yn3, ':')
plt.legend(['data', 'linear', 'cubic', 'spline', 'nearest'], loc = 'best')
plt.show()


#Discussão: Neste caso, a função plotada foi f(x)=sin(x)*sqrt(x) e mais uma vez, o comportamento dos 3 métodos foi
#semelhante, porém, o método "Cubic spline" se ajustou ligeiramente melhor à curva desejada, portando, parece o mais
#adequado para a situação.
```
